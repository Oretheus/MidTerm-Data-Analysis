## Import Libraries

```python
# Importing all necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Library for making a geospatial analysis
!pip install folium
import folium
```

## Cleaning data

- Before delving into the data analysis, the most crucial step is to clean the data. This process ensures the accuracy, reliability, and validity of the analysis. Clean data is fundamental to obtaining meaningful, actionable, consistent, and accurate insights. 

```python
def load_clean_uk_data(file_path):
    # Load the (UK) dataset
    df_uk = pd.read_csv(file_path)
    
    # Replace empty strings and 'NaN' with actual NaN values
    df_uk.replace("", np.nan, inplace=True)
    df_uk.fillna(np.nan, inplace=True)

    # Convert columns to appropriate data types if necessary
    df_uk['date'] = pd.to_datetime(df_uk['date'], format='%d/%m/%Y', errors='coerce')
    df_uk['time'] = pd.to_datetime(df_uk['time'], format='%H:%M', errors='coerce')

    return df_uk
```

```python
def load_clean_us_data(file_path):
    # Load the dataset
    df_us = pd.read_csv(file_path)
    
    # Replace empty strings and 'NaN' with actual NaN values
    df_us.replace("", np.nan, inplace=True)
    df_us.fillna(np.nan, inplace=True)

    # Convert columns to appropriate data types if necessary
    df_us['Start_Time'] = pd.to_datetime(df_us['Start_Time'], errors='coerce')
    df_us['End_Time'] = pd.to_datetime(df_us['End_Time'], errors='coerce')

    return df_us
```

*Creating variables for the filepaths for easier access*

```python
# File paths
uk_file_path = 'dft-road-casualty-statistics-collision-2022.csv'
us_file_path = 'US_Accidents_March23.csv'
```

## Accident Severity Analysis

```python
# Load the cleaned UK dataset
df_uk = load_clean_uk_data(uk_file_path)

# Plot the distribution of accident severity
plt.figure(figsize=(10, 6))
sns.countplot(x='accident_severity', data=df_uk, palette='viridis')
plt.title('Distribution of Accident Severity in UK')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.show()

# Load the cleaned US dataset
df_us = load_clean_us_data(us_file_path)

# Plot the distribution of accident severity
plt.figure(figsize=(10, 6))
sns.countplot(x='Severity', data=df_us, palette='viridis')
plt.title('Distribution of Accident Severity in US')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.show()
```

- The UK appears to have a higher rate of severe accidents compared to the US. This discrepancy can be attributed to several factors, most notably the contrasting road conditions between the two regions. In the UK, narrow roadways seem to be more common, which may contribute to a higher likelihood of severe accidents. Conversely, the area in the US under observation features wide, expansive roads and relatively flat terrain, which can facilitate safer driving conditions and potentially reduce the severity of accidents. 

- I did a quick online search and found that my analysis could be correct, even though it is limited to the specific areas I am observing. According to an article from Cruise America, a key difference between driving in the UK and the US is the nature of the roadways. The UK often has narrow, more confined roads, whereas the US typically boasts wider and more open roads. This difference in roadway design could significantly impact driving safety and accident severity, supporting the idea that road conditions play a crucial role in the variance of accident rates between these two regions [Link](https://www.cruiseamerica.com/trip-inspiration/differences-driving-in-uk-and-usa#:~:text=As%20a%20general%20rule%20of,difference%20in%20country%20size%20overall.).

<h1>Geospatial Analysis</h1>

```python
# UK Map Visualization
uk_map = folium.Map(location=[54.5, -4], zoom_start=6)

for index, row in df_uk.iterrows():
    folium.CircleMarker([row['latitude'], row['longitude']],
                        radius=3,
                        color='red' if row['accident_severity'] == 1 else 'orange' if row['accident_severity'] == 2 else 'green',
                        fill=True).add_to(uk_map)

uk_map.save('uk_accidents_map.html')

# US Map Visualization
us_map = folium.Map(location=[37.6, -95.665], zoom_start=4)

for index, row in df_us.iterrows():
    folium.CircleMarker([row['Start_Lat'], row['Start_Lng']],
                        radius=3,
                        color='red' if row['Severity'] == 4 else 'orange' if row['Severity'] == 3 else 'yellow' if row['Severity'] == 2 else 'green',
                        fill=True).add_to(us_map)

us_map.save('us_accidents_map.html')

# The maps are saved in the same directory folder if you would like to see them
```

- Geospatial Analysis is really useful and helpful for understanding accident severity because it helps you see the layout of different regions, roads, and landscapes. By creating maps and visual tools, it becomes easier to spot where accidents happen more often and why. This kind of analysis makes it clear how the physical setup of an area might influence the rate of severe accidents, helping to support other data and conclusions about the factors affecting accident severity.

# Temporal Analysis

```python
# Convert 'time' column to datetime and extract hour for UK data
df_uk['time'] = pd.to_datetime(df_uk['time'], format='%H:%M', errors='coerce')
df_uk['hour'] = df_uk['time'].dt.hour

# Check for NaNs in 'hour' after conversion
nan_times = df_uk[df_uk['hour'].isna()]
if not nan_times.empty:
    print("Rows with NaN hours in UK data:")
    print(nan_times[['time']])

# Convert 'Start_Time' column to datetime and extract hour for US data
df_us['Start_Time'] = pd.to_datetime(df_us['Start_Time'], errors='coerce')
df_us['hour'] = df_us['Start_Time'].dt.hour

# Prepare for line plot: count occurrences of each hour
uk_hour_counts = df_uk['hour'].value_counts().sort_index()
us_hour_counts = df_us['hour'].value_counts().sort_index()

# Plot for UK accidents
plt.figure(figsize=(12, 6))
if not uk_hour_counts.empty:
    sns.lineplot(x=uk_hour_counts.index, y=uk_hour_counts.values, color='blue', label='UK')
    plt.title('Accident Times in the UK')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Accidents')
    plt.grid(True)
    plt.xticks(range(0, 24))
    plt.legend()
else:
    plt.text(0.5, 0.5, 'No data available for UK', horizontalalignment='center', verticalalignment='center')
plt.show()

# Plot for US accidents
plt.figure(figsize=(12, 6))
if not us_hour_counts.empty:
    sns.lineplot(x=us_hour_counts.index, y=us_hour_counts.values, color='red', label='US')
    plt.title('Accident Times in the US')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Accidents')
    plt.grid(True)
    plt.xticks(range(0, 24))
    plt.legend()
else:
    plt.text(0.5, 0.5, 'No data available for US', horizontalalignment='center', verticalalignment='center')
plt.show()
```


**UK Accident Times:**

- In the UK, the majority of accidents occur in the afternoon. Particularly during the late afternoon hours. This finding is within expectation as the afternoon is popular for its rush hours, this is when people are returning from work or running errands, which seems to result in higher accident rates. The peak in accidents during the afternoon hours suggests a correlation between commuter traffic and accident frequency.

**US Accident Times:**

- Differently to the UK, the US exhibits a nearly similar pattern, with the highest number of accidents occurring in the morning. Particularly during the early morning. This pattern could be attributed to the same reasoning as the UK. Instead of afternoon rush hour traffic, it is on the contrary morning rush hour traffic. Potential drowsiness among drivers who have just woken up, and increased activity as people begin their day.

**Unexpected Patterns:**

- The difference in accident patterns between the two countries was somewhat unexpected. While the afternoon peak in the UK aligns with common expectations, the morning peak in the US was rather surprising. This unexpected finding highlights the importance of considering cultural and environmental factors when analyzing accident data.

**Data Limitations:**

**Research and Methodology:**

 - To address the limitation regarding the absence of day-of-the-week information, I decided to conduct further researc. It became evident that including the day of the week in the analysis could offer a more comprehensive understanding of accident patterns. Therefore, an update was made to the analysis code to incorporate this additional variable using heatmaps.

```python
# Create a new DataFrame to aggregate accidents counts by day of the week and hour
uk_accidents_by_day_hour = df_uk.groupby(['day_of_week', 'hour']).size().unstack(fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(uk_accidents_by_day_hour, cmap='YlGnBu')
plt.title('Accident Distribution by Day of Week and Hour (UK)')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create a new DataFrame to aggregate accidents counts by day of the week and hour for US data
us_accidents_by_day_hour = df_us.groupby(['day_of_week', 'hour']).size().unstack(fill_value=0)

# Plot the heatmap for US data
plt.figure(figsize=(12, 8))
sns.heatmap(us_accidents_by_day_hour, cmap='YlGnBu')
plt.title('Accident Distribution by Day of Week and Hour (US)')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Weather Conditions Analysis

```python
# Mapping numerical codes to weather conditions for the UK dataset
weather_mapping_uk = {
    1: 'Clear or Sunny',
    2: 'Partly Cloudy',
    3: 'Overcast',
    4: 'Rainy',
    5: 'Snowy',
    6: 'Foggy',
    7: 'Windy',
    8: 'Other/Unknown'
}

# Replace numerical codes with corresponding weather conditions
df_uk['weather_conditions'] = df_uk['weather_conditions'].map(weather_mapping_uk)

# Plot the distribution of weather conditions for UK dataset
plt.figure(figsize=(12, 6))
sns.countplot(y='weather_conditions', data=df_uk, order=df_uk['weather_conditions'].value_counts().index, palette='viridis')
plt.title('Accidents by Weather Conditions in UK')
plt.xlabel('Count')
plt.ylabel('Weather Conditions')
plt.show()

# Plot the distribution of weather conditions for UK dataset
plt.figure(figsize=(12, 6))
sns.countplot(y='Weather_Condition', data=df_us, order=df_us['Weather_Condition'].value_counts().index, palette='viridis')
plt.title('Accidents by Weather Conditions in US')
plt.xlabel('Count')
plt.ylabel('Weather Conditions')
plt.show()
```

**Prevalence of Clear Days:**

- Contrary to initial expectations, a significant proportion of accidents occurred on clear or sunny days. This finding was unexpected, as one might assume that adverse weather conditions, such as rain or snow, would lead to more accidents. However, it appears that weather plays a lesser role in accident occurrence than anticipated.

**Impact of Data Size:**

- Again, it's important to acknowledge the limitations of the analysis due to the size of the dataset. With a larger dataset, we may obtain a more accurate representation of the relationship between weather conditions and accidents. The current findings seem to be skewed and incomplete. Going off of pure intellect, I'm sure that weather conditions have a more severe impact than what has been found here.

**Other Weather Conditions:**

- While clear days accounted for a significant portion of the accidents, it's worth noting that accidents also occurred under various other weather conditions, including partly cloudy, overcast, rainy, snowy, foggy, and windy conditions. This diversity suggests that weather conditions, although not the primary factor, still contribute to accident occurrence to some extent. Maybe in the future I can expound on this with a larger dataset.

```python
# Replace numerical codes with descriptive labels for road surface conditions in the UK dataset
df_uk['road_surface_conditions'].replace({
    1: 'Dry',
    2: 'Wet/Damp',
    3: 'Snow',
    4: 'Frost/Frozen',
    5: 'Flood (Surface Water Over 3 cm Deep)',
    6: 'Oil/Diesel',
    7: 'Mud',
    8: 'Other',
    9: 'Unknown'
}, inplace=True)

# Plot the distribution of road surface conditions for UK dataset
plt.figure(figsize=(12, 6))
sns.countplot(y='road_surface_conditions', data=df_uk, order=df_uk['road_surface_conditions'].value_counts().index, palette='viridis')
plt.title('Accidents by Road Surface Conditions in UK')
plt.xlabel('Count')
plt.ylabel('Road Surface Conditions')
plt.show()

plt.subplot(1, 2, 2)
sns.countplot(y='Bump', data=df_us, order=df_us['Bump'].value_counts().index, palette='magma')
plt.title('Road Surface Conditions in US')
plt.xlabel('Count')
plt.ylabel('Surface Conditions')

plt.tight_layout()
plt.show()
```

**Analysis:**

- For the UK dataset, the road surface conditions provide valuable insights into the circumstances surrounding accidents. The majority of accidents occurred on dry roads, which is not unexpected given typical weather conditions. However, it's noteworthy that a significant number of accidents also occurred under wet or damp road conditions, indicating the increased risk of accidents during adverse weather. Additionally, there were occurrences of accidents on frost/frozen roads, highlighting the challenges posed by winter weather conditions. 

- Unfortunately, the US dataset did not have an equivalent road surface condition variable. However, it's interesting to note that not a single accident in the US dataset was attributed to bumps. This observation suggests that road surface irregularities, such as bumps, may not be significant contributors to accidents in the US dataset. Instead, the absence of bumps as a contributing factor implies that accidents may predominantly result from human error or other environmental factors.
